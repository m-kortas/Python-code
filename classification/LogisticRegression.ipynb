```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Imputer
```

```python
pd.options.mode.chained_assignment = None # default='warn'  nie pokazuje błędu
```

```python
X = train[["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]]  # data frame

X["Sex"] = X["Sex"].astype("category") # kategoria
X["Sex"] = X["Sex"].cat.codes  # kody kategorii

X["Embarked"] = X["Embarked"].astype("category")  #kategoria
X["Embarked"] = X["Embarked"].cat.codes  # kody kategorii
```

```python
from sklearn.preprocessing import Imputer
imputer = Imputer(strategy = "median")   #wrzucanie mediany
X = pd.DataFrame(imputer.fit_transform(X), columns = X.columns, index = X.index) #NA na mediany
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()     #zapamiętywanie strategii
```

```python
X = scaler.fit_transform(X.values) #skaluje dane
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression() ## classifier
```

```python
y = train["Survived"] #Survived
```

```python
classifier.fit(X,y) #Classifier z Survived FIT
```

```python
train["Predicted"] = classifier.predict(X) #predycja
```
